```python
# Necessary imports
import pandas as pd
import csv

# Read in Master VCF file
df = pd.read_csv(r"final_MASTERVCF.txt", sep="\t", comment='#')

# Convert to pd df to list 
GENE_list = df['GENE'].tolist()

# Iterate through list and count each gene through dictionary
GENE_dict = {}
for gene in GENE_list:
    if gene in GENE_dict:
        # Increment the count if the item is already present
        GENE_dict[gene] += 1
    else:
        # Add the item to the dictionary with a count of 1 if not present
        GENE_dict[gene] = 1
```

```python
# Specify the CSV file name
csv_file_name = 'gene_count.csv'

# Write dictionary as a CSV file
with open(csv_file_name, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    # Write each key-value pair as a new row
    for key, value in GENE_dict.items():
        writer.writerow([key, value])

print(f"CSV file '{csv_file_name}' has been created.")
```
